```python
# Working with date-time data
df = pd.read_csv('./data/ETH_1h.csv')
df.head()
```

```python
df['Date'] =  pd.to_datetime(df['Date'],
                             format = '%Y-%m-%d %I-%p')
df['Date']
```

```python
df['Date'].dt.day_name()
```

```python
# Filter dates
filt = (df['Date'] >= '2020')
df.loc[filt]
```

```python
filt = (df['Date'] >= '2019') & (df['Date'] < '2020')
df.loc[filt]
```

```python
# Alternatively
filt = (df['Date'] >= pd.to_datetime('2019-01-01')) & (df['Date'] < pd.to_datetime('2020-01-01'))
df.loc[filt]
```
